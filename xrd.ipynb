```python
#déterminer les 3 premiers pics du diffractogramme correspondant (en indiquant les indices hkl associés) 
#pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å.
```

```python
from pymatgen.ext.matproj import MPRester 
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
#structure
with MPRester("TJVQZXfS2KHqvsC7QEip") as m:
    structure = m.get_structure_by_material_id("mp-505297")
```

```python
#diffractogramme 
xrd = XRDCalculator('CuKa1')
xrd = XRDCalculator(wavelength = 'CuKa', symprec = 0, debye_waller_factors = None)
#plot de tous les pics 
print('plot de tous les pics')
xrd.show_plot(structure)
#informations sur les trois premiers pics : 
xrdpattern = xrd.get_pattern(structure = structure, scaled = True, two_theta_range = (0,90))
for i in range(0,3) :
    print("Pic de diffraction ", i+1 , " : 2 \u0398 = ", xrdpattern.x[i], 'degrés')
    print("                       Intensité (sur une échelle de 100) : ", xrdpattern.y[i]) 
    print("                       Indices hkl : ", xrdpattern.hkls[i][0]['hkl'])
    print("                       multiplicité : ", xrdpattern.hkls[i][0]['multiplicity'])
    print('\n')
```
